# 繪製 BBands

```python
%matplotlib inline
import pandas as pd
from datetime import datetime
import pandas_datareader.data as web
```

```python
# 讀取從指定日期之後的股價資訊
df = web.DataReader("3008.tw", "yahoo", datetime(2016,1,1))
```

## 畫出BBands

```python
# bbands策略,N=20
df['20ma'] = pd.Series.rolling(df['Adj Close'], window=20).mean()
df['std'] = pd.Series.rolling(df['Adj Close'], window=20).std()
# 上軌=20ma+1.5sd ,中軌=20ma, 下軌=20ma-1.5sd
df['ubb'] = df['20ma'] + 1*df['std']
df['mbb'] = df['20ma']
df['lbb'] = df['20ma'] - 1*df['std']
```

```python
df[['Adj Close','ubb','mbb', 'lbb']].plot(grid=True, figsize=(20,10))
```
