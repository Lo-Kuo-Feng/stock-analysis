````python
from datetime import datetime
import pandas_datareader.data as web
````

## 設定資料的起始範圍

````python
start = datetime(2016, 1, 1)
end = datetime(2018, 12, 31)
````

## 讀取個股歷史資料

````python
df = web.DataReader('3008.tw', "yahoo", start, end)
````

````python
df.info()
````

````python
df['2018-6-1':'2019-2-10']
````
